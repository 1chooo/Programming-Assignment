```python
import wfdb
import os
import matplotlib.pyplot as plt

record_name = 'tr03-1010'
record_path = os.path.join(train_path, record_name, record_name)
annotation = wfdb.rdann(record_path, 'arousal')

# Get the indices and symbols for the sleep stages
stage_indices = annotation.sample
stage_symbols = annotation.symbol

# Create a dictionary mapping the stage symbols to their respective numbers
stage_mapping = {
    'N1': 1,
    'N2': 2,
    'N3': 3,
    'R': 5,
    'W': 4
}

# Filter the indices and symbols to include only the desired sleep stages
filtered_indices = [i for i, symbol in enumerate(stage_symbols) if symbol in stage_mapping.keys()]
filtered_symbols = [stage_symbols[i] for i in filtered_indices]
filtered_samples = [stage_indices[i] for i in filtered_indices]

# Plot the sleep stages
plt.figure(figsize=(12, 4))
plt.plot(filtered_samples, [stage_mapping[symbol] for symbol in filtered_symbols], 'ro', markersize=4)
plt.title('Sleep Stages')
plt.xlabel('Time')
plt.ylabel('Stage')
plt.yticks([1, 2, 3, 4, 5], ['N1', 'N2', 'N3', 'W', 'R'])
plt.grid(True)
plt.show()
```
